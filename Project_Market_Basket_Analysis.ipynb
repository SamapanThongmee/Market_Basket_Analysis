##### **Data Preparation for Apriori Algorithm**

```python
transactions=[]
for item in data['Transaction'].unique():
    lst=list(set(data[data['Transaction']==item]['Item']))
    transactions.append(lst)

transactions[0:20]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

transaction_encoder = TransactionEncoder()
encoded_data = transaction_encoder.fit(transactions).transform(transactions)
apriori_data = pd.DataFrame(encoded_data, columns=transaction_encoder.columns_).astype(int)
apriori_data.head()
```

##### **Association Rules**

```python
frequent_items= apriori(apriori_data, use_colnames=True, min_support=0.02)
frequent_items.head()
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.50)
rules.antecedents = rules.antecedents.apply(lambda x: next(iter(x)))
rules.consequents = rules.consequents.apply(lambda x: next(iter(x)))
rules = rules[['antecedents', 'consequents', 'antecedent support', 'consequent support', 'support', 'confidence', 'lift']].rename(
    columns={'antecedents':'Antecedents', 'consequents':'Consequents', 'antecedent support':'Antecedent Support', 'consequent support':'Consequent Support', 'support':'Support', 'confidence':'Confidence', 'lift':'Lift'}
)
rules
```

```python
Filter_rules = rules[(rules['Support'] >= 0.03) & (rules['Confidence'] >= 0.50)]
Filter_rules

# https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/#association_rules-association-rules-generation-from-frequent-itemsets
```
